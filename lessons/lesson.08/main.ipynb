```python
class Dictable:
    def to_dict(self):
        return {
            key: value
            for key, value in self.__dict__.items()
            if not key.startswith("_")
        }
```

```python
class User(Dictable):

    def __init__(self, name, age):
        self.name = name
        self.age = age

    def describe(self):
        print("I'm", self.name, self.age, "years old")

    def __str__(self):
        return f"{self.name=}, {self.age=}"
```

```python
class Friendable:
    def __init__(self):
        self._friends_joined = []

    def add_friend(self, other):
        self._friends_joined.append(other)
        other._friends_joined.append(self)
```

```python
class User(Serializable, Friendable):

    def __init__(self, name, age):
        self.name = name
        self.age = age
        self._secret = "something personal"

    def describe(self):
        print("I'm", self.name, self.age, "years old")

    def set_secret(self, secret):
        self._secret = secret
    
    def tell_secret(self):
        print("My secret is:", self._secret)

    def __str__(self):
        return f"{self.name=}, {self.age=}"
```

```python
import json


class Serializable(Dictable):
    def __init__(self):
        print("inited Serializable")

    def to_json(self):
        return json.dumps(self.to_dict())
```

```python
class User(Serializable, Friendable):

    def __init__(self, name, age):
        super().__init__()
        self.name = name
        self.age = age
        self._secret = "something personal"
        print("inited User")

    def describe(self):
        print("I'm", self.name, self.age, "years old")

    def set_secret(self, secret):
        self._secret = secret
    
    def tell_secret(self):
        print("My secret is:", self._secret)

    def __str__(self):
        return f"{self.name=}, {self.age=}"
```

```python
class User(Serializable, Friendable):

    def __init__(self, name, age):
        Serializable.__init__(self)
        Friendable.__init__(self)
        self.name = name
        self.age = age
        self._secret = "something personal"
        print("inited User")

    def describe(self):
        print("I'm", self.name, self.age, "years old")

    def set_secret(self, secret):
        self._secret = secret
    
    def tell_secret(self):
        print("My secret is:", self._secret)

    def __str__(self):
        return f"{self.name=}, {self.age=}"

    def __repr__(self):
        return str(self)
```
