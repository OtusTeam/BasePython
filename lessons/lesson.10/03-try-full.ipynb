```python
from timeit import default_timer as timer


def div_safe(a, b):
    start_time = timer()
    result = a / b
    end_time = timer()
    print(a, "/", b, "=", result)
    total_time = end_time - start_time
    print(f"got result in {total_time:.9f}s")
    return result
```

```python
def div_safe(a, b):
    start_time = timer()
    try:
        result = a / b
    except ZeroDivisionError:
        print("division by zero, returning 0")
        return 0
    else:
        print(a, "/", b, "=", result)
        return result
    finally:
        end_time = timer()
        total_time = end_time - start_time
        print(f"got result in {total_time:.9f}s")
```

```python
div_safe(1, 2)
```

```python
div_safe(12345, 24681)
```
