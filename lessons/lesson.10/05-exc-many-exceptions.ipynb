```python
def div_safe(a, b):
    try:
        return a / b
    except ZeroDivisionError:
        print("Division by zero, returning 0")
        return 0
    except TypeError as e:
        print("invalid types:", e, "returning None")
        return None
```

```python
div_safe(1, 2)
```

```python
div_safe(1, 0)
```

```python
div_safe("1", 1)
```
