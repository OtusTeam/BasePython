```python
from dataclasses import dataclass


@dataclass
class User:
    id: int
    username: str
    full_name: str
    age: int

    def increase_age(self):
        self.age += 1
        return self.age
```

```python
john = User(id=1, username="john", full_name="John Smith", age=22)
```

```python
john
```

```python
john.increase_age()
```

```python
john
```

```python
@dataclass(frozen=True)
class Food:
    name: str
    weight: int
    price: int
```

```python
bread = Food(name="bread", weight=500, price=100)
bread
```

```python
bread.weight
```

```python
bread.price
```

```python
bread.price *= 2
```

```python
from dataclasses import field


@dataclass(slots=True)
class Comment:
    text: str
    entities: list[tuple[int, int, str]] = field(default_factory=list)
```

```python
comment = Comment(text="This is the comment")
comment
```

```python
comment.text = "hello world"
comment
```

```python
comment.foobar = "qwerty"
```

```python
comment.__slots__
```
