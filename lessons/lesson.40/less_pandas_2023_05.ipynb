```python
def age_new(x):
    if x > 40:
        return 'Пенсионер'
    return 'Молодой'

df['MyCol'] = df['Age'].apply(age_new)
```

```python
def age_new1(row):
    if row['Age'] >= 40 and row['Gender'] == 'Male':
        return 'Пенсионер'
    elif row['Age'] >= 40 and row['Gender'] == 'Female':
        return 'Пенсионерка'
    elif row['Age'] < 40 and row['Gender'] == 'Male':
        return 'Молодой'
    return 'Молодуха'
```

```python
df['MyCol1'] = df.apply(age_new1, axis=1)
```
