```python
def convert_memory(x):
    memory = x.split()[0]
    if memory.endswith('GB'):
        res = memory.replace('GB', '')
    elif memory.endswith('TB'):
        res = float(memory.replace('TB', '')) * 1024
    return int(res)
        
    
```

```python
df['Memory_amount'] = df['Memory'].apply(convert_memory)
```

```python
df.head()
```

```python
df.info()
```

```python
def convert_(row):
    return int(row['Weight1'] * 100 + row['Memory_amount'])
```

```python
df['convert1'] = df.apply(convert_, axis=1)
```
