```python
def convert_to_memory(x):
    memory = x.split()[0]
    if memory.endswith('GB'):
        res = memory.replace('GB', '')
    elif memory.endswith('TB'):
        res = float(memory.replace('TB', '')) * 1024
    return int(res)
        
```

```python
df1['Memory_amount'] = df['Memory'].apply(convert_to_memory)
```
